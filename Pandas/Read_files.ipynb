### Writing Csv

```python
df = pd.read_csv(
    'bookdata.csv', 
    nrows = 100 , 
    usecols = [0,1,3,4,5,7,8], 
    dtype= {'availability': 'float64'},
    na_values= ['sequential art', 'self help']
)
df.tail(8)
```

```python
df.isnull().sum()
```

```python
df.dropna(axis = 1)
```

```python
df.dropna(subset = ['category']).tail(7)
```
